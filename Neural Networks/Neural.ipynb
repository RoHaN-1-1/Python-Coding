```python
inp = [1,2,3,2.5]
weight = [[0.2,0.8,-0.5,1],
      [0.5,-0.91,0.26,-0.5],
      [-0.26,-0.27,0.17,0.87]]
bias = [2,3,0.5]
output_layer = []

for i in range(len(weight)):
    out = 0

    for j in range(len(inp)):
        out += inp[j] * weight[i][j]

    out += bias[i]
    output_layer.append(out)

print(output_layer)
```

```python
import numpy as np
inp = [1,2,3,2.5]
weight = [[0.2,0.8,-0.5,1],
      [0.5,-0.91,0.26,-0.5],
      [-0.26,-0.27,0.17,0.87]]
bias = [2,3,0.5]
output_layer = np.dot(weight,inp) + bias

print(output_layer)
```
